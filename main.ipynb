```python
import pandas as pd
```

```python
df=pd.read_csv('./FDI_in_India.csv')
```

```python
col = ['sector','2000', '2001', '2002', '2003', '2004', '2005','2006', '2007', '2008',
  '2009', '2010', '2011','2012', '2013', '2014', '2015', '2016']
```

```python
df.columns= col
```

```python
df
```

```python
yearly_fdi=df[['2000','2001','2002', '2003', '2004', '2005','2006', '2007', '2008',
   '2009', '2010', '2011','2012', '2013', '2014', '2015', '2016']].sum()

sort_yearly_fdi=yearly_fdi.sort_values(ascending=False)
sort_yearly_fdi
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14, 6))
plt.bar(sort_yearly_fdi.index,sort_yearly_fdi.values)
plt.xlabel('YEAR')
plt.ylabel('AMOUNT $')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.plot(yearly_fdi.index,yearly_fdi.values)
plt.show()
```

```python
total_fdi_bysector=df[['2000','2001','2002', '2003', '2004', '2005','2006', '2007', '2008',
   '2009', '2010', '2011','2012', '2013', '2014', '2015', '2016']].sum(axis=1)
total_fdi_bysector.index=df['sector']
total_fdi_bysector=total_fdi_bysector.sort_values(ascending=False)
print('TOP 5 SECTORS WITH HIGHEST FDI')
total_fdi_bysector.head()
```

```python
# UNPIVOTING TABLE

years=['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008',
       '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016']
sectors=['sector']

dfr=pd.melt(df, id_vars = sectors, value_vars = years, var_name='Year',value_name='FDI(million $)',ignore_index=True)
dfr
```
